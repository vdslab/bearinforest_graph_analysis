```python
import re

# タイトルの前処理
# texコマンドっぽいのがあったのでそれの除去
def deleteTexCommand(str):
    # tagの除去
    reg_obj = re.compile(r"<[^>]*?>")
    new_str = reg_obj.sub("", str)
    # print(j_title)
    # $の除去
    new_str = new_str.replace("$", "")
    # print(new_str)

    return new_str
```

```python

# titleとabstractの取得
def getTitleAbstract(article_object_array):
    articles = []
    # title, abstract部分だけ取り出す
    for i, item in enumerate(article_object_array):
        # 著者の取得
        authors = item["authors"]["authors"]
        # print("authors:{}".format(authors))
        # title，abstract，著者の情報がある場合
        if("title" in item and "abstract" in item and authors):
            # print("id:{} title {}".format(i, item["title"]))
            # titleの取得
            title = deleteTexCommand(item["title"])
            # abstractの取得
            abstract = item["abstract"]
            articles.append([title, abstract])
        # else:
            # 欠落した情報の表示
            # if("title" not in item):
            #     print("not found title")
            # if("abstract" not in item):
            #     print("not found abstract")
            # if(not authors):
            #     print("not found authors")
            # print("\tin {}".format(i))

    return articles
```

```python
import os
import glob
import pprint

def getPublicNmaes(paper_directory):
  dir = paper_directory
  if(paper_directory[-1] != "/"):
    dir += "/"
  # 2020のデータまでパス
  filepath = glob.glob(dir + "*/2020*.json")
  # pprint.pprint(filepath)

  dirname = []
  for path in filepath:
    # そのデータの直上のファイル名を取得
    subdirname = os.path.basename(os.path.dirname(path))
    # print(subdirname)
    dirname.append(subdirname)
    # print(os.path.dirname(filepath[0]))

  # 重複をなくす
  dirname = list(set(dirname))
  pprint.pprint(dirname)
  return dirname
```

```python
import json
import csv
import glob
import pprint
import os


# 雑誌のディレクトリのパス
public_dir = "/content/drive/Shareddrives/vdslab/2022/bearinforest/data/"
public_titles = getPublicNmaes(public_dir)
# csvを保存したいディレクトリのパス
# 存在しなくても作ってくれるよ
write_csv_dir = "/content/drive/MyDrive/zemi_3/doc2vec/title_abst_csv"

for public_title in public_titles:
    # フォルダ作成
    if not os.path.isdir(write_csv_dir):
        os.makedirs(write_csv_dir)
    # ファイルの初期化
    with open(write_csv_dir + '/' + public_title + '.csv', 'w', newline="") as f:
        writer = csv.writer(f)
        writer.writerow(['title', 'abstract'])

    for year in range(1900, 2022):
        print(year)
        # jsonファイルのパスを取得
        files = glob.glob(public_dir + public_title + "/" + str(year) + "*.json")
        # print(files)

        for file_path in files:
            # print(file_path)
            s = os.path.getsize(file_path)
            if(s <= 0): continue
            json_read = open(file_path, "r")
            # jsonファイルとして読み込む
            json_file = json.load(json_read)
            # 読み込みの終了
            json_read.close()

            # titleとabstractの取得
            articles = getTitleAbstract(json_file["articles"])
            # pprint.pprint(articles[0:3])

            # title, abstractをcsvファイルに書き込み
            with open(write_csv_dir + '/' + public_title + '.csv', 'a', newline="") as f:
                writer = csv.writer(f)

                print(len(articles))
                for item in articles:
                    writer.writerow([item[0], item[1]])
```
