```python
import os
import glob
import pprint

def getPublicNmaes(paper_directory):
  dir = paper_directory
  if(paper_directory[-1] != "/"):
    dir += "/"
  # 2020のデータまでパス
  filepath = glob.glob(dir + "*/2020*.json")
  # pprint.pprint(filepath)

  dirname = []
  for path in filepath:
    # そのデータの直上のファイル名を取得
    subdirname = os.path.basename(os.path.dirname(path))
    # print(subdirname)
    dirname.append(subdirname)
    # print(os.path.dirname(filepath[0]))

  # 重複をなくす
  dirname = list(set(dirname))
  pprint.pprint(dirname)
  return dirname
```

```python
import pprint
from gensim.models.doc2vec import Doc2Vec
from gensim.models.doc2vec import TaggedDocument
import csv
import time

# 雑誌のディレクトリのパス
public_dir = "/content/drive/Shareddrives/vdslab/2022/bearinforest/data/"
public_titles = getPublicNmaes(public_dir)
# titleとabstractのcsvがあるディレクトリのパス
csv_dir = "/content/drive/MyDrive/zemi_3/doc2vec/title_abst_csv/"

model_save_dir = "/content/drive/MyDrive/zemi_3/doc2vec/"
model_name = "doc2vec_model"
print("model name >> {}".format(model_name))

time_start = time.time()

print("***** start make doc2vec model *****")
print("get train data")
train_csv_data = []
for title in public_titles:
    print("\topen {}".format(title))
    with open(csv_dir + title + '.csv', encoding="utf-8") as f:
        reader = csv.reader(f)
        csvdata = [row for row in reader]
        csvdata = csvdata[1:]
        # print(csvdata)
        # abstract:1, title:0
        train_csv_data.extend(csvdata)
# print(train_csv_data[0:3])
train_data = ([TaggedDocument(words=train_csv_data[i][1], tags=[
    train_csv_data[i][0]]) for i in range(len(train_csv_data))])

# pprint(train_csv_data)
# print(train_data)
print("train data length {}".format(len(train_data)))

print("start train")
time_train_start = time.time()

print(model_name)
model = Doc2Vec(documents=train_data)
print("save")
model.save(model_save_dir + "{}.model".format(model_name))

time_train_end = time.time()
print("time {}\n".format(time_train_end - time_train_start))

time_end = time.time()
print(time_end - time_start)
print("***** end make doc2vec model *****")
```
